```python
def extract_values(str_lst):
    values = json.loads(str_lst)
    return [i["name"] for i in values if i["job"]=="Director"]
```

```python
def collapse(lst):
    new_lst = []
    for i in lst:
        new_lst.append(i.replace(" ",""))     #removes empthy spaces b/w names
    return new_lst
```

### Stem words

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
```

```python
ps = PorterStemmer()
```

```python
ps.stem("beauty")
```

```python
ps.stem("beautiful")
```

```python
def stem(txt):
    lst = []
    for i in txt.split():          #every word must be splitted first and converted into list and then stemmed
        lst.append(ps.stem(i))
    
    return " ".join(lst)
```

```python
stem(final_df["tags"][0])
```

```python
final_df["tags"] = final_df["tags"].apply(stem)
```

## Recommendation

```python
movie_index = final_df[final_df["title"] == "Iron Man"].index[0]  #index of movie
movie_index
```

```python
distances = similarity[movie_index]                #similarity of movie with other movies
distances
```

```python
movie_index = list(enumerate(distances))
movie_index

#enumerate() returns list of similarity along with indeces
```

```python
similar_movie = sorted(movie_index, reverse =True, key=lambda x: x[1])[1:11]
similar_movie
#returns top 10 movie with highest similarity
#func in lambda ignores index no. and sorts acc to similarity
```

```python
for index, i in similar_movie:
    print(final_df.iloc[index, 1])
```

```python
def recommender(movie_name):
    movie_index = final_df[final_df["title"] == movie_name].index[0]
    distances = similarity[movie_index]
    movie_index = list(enumerate(distances))
    similar_movie = sorted(movie_index, reverse =True, key=lambda x: x[1])[1:11]
    for index, i in similar_movie:
        print(final_df.iloc[index, 1])
```

```python
def fetch_poster(movie_id):
    url = f"https://api.themoviedb.org/3/movie/{movie_id}?api_key=75eb0685f1f9140663e33eb0ea57150a&language=en-US"
    data = requests.get(url)
    data = data.json()
    poster_path = data['poster_path']
    full_path = "https://image.tmdb.org/t/p/w500/" + poster_path
    return full_path

#fetches poster of desired movie id
```

```python
fetch_poster(19995)
```

```python
def recommender(movie_name):
    movie_index = final_df[final_df["title"] == movie_name].index[0]
    distances = similarity[movie_index]
    movie_index = list(enumerate(distances))
    similar_movie = sorted(movie_index, reverse =True, key=lambda x: x[1])[1:11]
    
    recommended_movies = []
    recommended_movies_poster = []
    
    for index, i in similar_movie:
        recommended_movies.append((final_df.iloc[index, 1]))
        movie_id = final_df.iloc[index, 0]
        recommended_movies_poster.append(fetch_poster(movie_id))
        
    return recommended_movies, recommended_movies_poster

#Returns recommended movies list along with the list of thier posters' url
```

```python
recommender("Iron Man")
```
